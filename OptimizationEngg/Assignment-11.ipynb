```python
from __future__ import division
import numpy as np
import sympy as sp
```

```python
a = np.linspace(-4,4, 100)
b = np.linspace(-4,4, 100)
```

```python
x1 = sp.Symbol('x1')
x2 = sp.Symbol('x2')
```

```python
fnc1 = (x1-1)**2 + x2**2
fnc2 = (x1+1)**2 + x2**2

eval_fnc1 =  np.array([fnc1.evalf(subs={x1:x, x2:y}) 
                                  for x in a for y in b], dtype='float32')
```

```python
eval_fnc2 =  np.array([fnc2.evalf(subs={x1:x, x2:y}) 
                                  for x in a for y in b], dtype='float32')
```

```python
eval_fnc1_eq = eval_fnc1[eval_fnc1<=2] 
a_eq = a[eval_fnc1<=2]
```
