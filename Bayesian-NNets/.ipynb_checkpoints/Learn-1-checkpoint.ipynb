```python
%matplotlib inline
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from scipy.stats import norm

sns.set_style('white')
sns.set_context('talk')

```

```python
# For simplicity we assume a normal distribution of mean(mu) piror  
# mean(mu) ~ Normal(0,1) and Liklihood (x|w ~ Normal(mu=0, std=1)
# So our posterior becomes P(w|x) = p(w) * p(x|w) / p(x)
# Here w are the weight parameters, which in this case are the mu and sigma

# Now that the priors of the mu and sigma are conjugate which means that the proterior distribution would take the
# same distribution as that of the priors
# Anlytical calculation of posterior for the mu and sigma:
```

```python
# Generate and vizualize data:
x = np.linspace(-1,1,500)
np.random.seed(123)
# Generate random 20 numbers
observed_data = np.random.randn(20)
print (observed_data)
ax = plt.subplot()
sns.distplot(observed_data, kde=False, ax=ax)
_ = ax.set(title='Histogram of observed data', xlabel='x', ylabel='# observations');
```

```python
# Prior using parameters mu and sigma, where mu=0 and sigma=1.
mu_prior = 0
sigma_prior = 1
ax = plt.subplot()
prior_dist = norm(mu_prior, sigma_prior).pdf(x)
ax.plot(x, prior_dist)
ax.set(xlabel='Linspace', ylabel='belief', title='Prior PDF distribution N~(0,1)');
sns.despine()
    
```

```python
# Likelihood using parameters mu and sigma using the observed data.
# When We have some data points 
print (observed_data)  # Lets say our data is generater from a random sample

mu_likelihood = np.mean(observed_data)
sigma_likelihood = np.std(observed_data)

print ('The parameter mu from the given data is: ', mu_likelihood)
print ('The parameter sigma from the given data is: ', sigma_likelihood)

likelihood = norm(mu_likelihood, sigma_likelihood).pdf(x)
ax = plt.subplot()
ax.plot(x, likelihood)
ax.set(xlabel='Data points', ylabel='belief', title='Likelihood PDF distribution N~(data_mean,data_std)');
sns.despine()
```

```python
# Posterior distribution is calculated by the use of prior and likelihood.
# In the 2nd plot we have a pdf distribution using parameters (mu=0 and sigma=1)
# In the 3rd plot we have a pdf distribution using parameters fetched from the data (mu=0.114417731955 and sigma=1.22561326874)
def call_posterior_analytical(data, mu_0, sigma_0):
    sigma = 1.   # Taking a constant variance for the "data"
    n = len(data)
    """
        Since the priors are conjugate the posterior has the same distribution as that of the prior. Using maximum 
        likelihood estimation and prior we have them as below.

        posterior P(mu|data) = P(mu|mu_0,sigma_0) * P(x1,x2,,xn|mu, sigma) / normalizer
        since we are using Gaussiam distribution,
        
        P(mu|mu_0,sigma_0) = 1/sq(sigma_0)*exp(-1/2*sq(sigma_0)sq(mu - mu_0)), and
        P(x1,x2,,xn|mu, mu_0, sigma_0) = 1/(sigma_0)**n * exp(-1/2*sq(sigma)sum(sq(mu - mu_0)))
        
        Similarly we get the posterior for sigma.
    
        using Schur complement of matrix we get the below simplified version of prior, likelihood and posterior
    """
    mu_prior = (sigma_0**2/(sigma**2 + sigma_0**2)) * mu_0           # we have sigma_0 and mu_0 leading because they are the prior belief
    mu_liklihood = (sigma**2/(sigma**2 + sigma_0**2)) * np.sum(data)
    mu_normalizer =  (1. / sigma_0**2) + (n / sigma**2)
    
    mu_post = mu_prior + mu_liklihood / mu_normalizer
    sigma_post = (1. / sigma_0**2 + n / sigma**2)**-1

#     print (mu_post)
#     print ('')
#     print (sigma_post)
    return mu_post, sigma_post #norm(mu_post, np.sqrt(sigma_post)).pdf(x)

mu_post, sigma_post = call_posterior_analytical(observed_data, 0., 1.)
posterior = norm(mu_post, np.sqrt(sigma_post)).pdf(x)
ax = plt.subplot()
ax.plot(x, posterior)
ax.set(xlabel='Data points', ylabel='belief', title='Posterior PDF distribution N~(data_mean,data_std)');
sns.despine()

```

```python
def plot_proposal(mu_current,mu_new,mu_prior,sigma_prior,data,
                  posterior_prob,accepted,mu_posterior,sample_no):
    from copy import copy
    mu_posterior = copy(mu_posterior)
    fig, (ax1, ax2, ax3, ax4) = plt.subplots(ncols=4, figsize=(16, 4))
    fig.suptitle('Iteration %i' % (sample_no + 1))
    x = np.linspace(-3, 3, 5000)
    
    if accepted:
        color = 'g' 
    else:
        color = 'r' 
        
    
    # Plot prior
    prior_current = norm(mu_prior, sigma_prior).pdf(mu_current) 
    prior_new = norm(mu_prior, sigma_prior).pdf(mu_new)
    prior = norm(mu_prior, sigma_prior).pdf(x)  
    # get the prior distribution using prior parameters for the linspace x
    ax1.plot(x, prior)  # Plot the probability density of the prior distribution
    ax1.plot([mu_current] * 2, [0, prior_current], marker='o', color='b') # Plot the Vertical blue line
    ax1.plot([mu_new] * 2, [0, prior_new], marker='o', color=color) #  Plot the Vertical Red line
    ax1.annotate("", xy=(mu_new, 0.2), xytext=(mu_current, 0.2), arrowprops=dict(arrowstyle="->", lw=2.))
    ax1.set(ylabel='Probability Density', 
            title='current: prior(mu=%.2f) = %.2f\nnew: prior(mu=%.2f) = %.2f' % (mu_current, prior_current, mu_new, prior_new))
    
    # Likelihood
    likelihood_current = norm(mu_current, 1).pdf(data).prod() 
    # get the likelihood of the data given the current mean (mu_current)
    likelihood_new = norm(mu_new, 1).pdf(data).prod()
    # get the likelihood of the data given the new mean (mu_mean)
    y_new = norm(loc=mu_new, scale=1).pdf(x)  
    # # get the likelihood pdf of the data given the new mean (mu_mean)
    sns.distplot(data, kde=False, norm_hist=True, ax=ax2)
    ax2.plot(x, y_new, color=color)
    ax2.axvline(mu_current, color='b', linestyle='--', label='mu_current')
    ax2.axvline(mu_new, color=color, linestyle='--', label='mu_new')
#     #ax2.title('Proposal {}'.format('accepted' if accepted else 'rejected'))
    ax2.annotate("", xy=(mu_new, 0.2), xytext=(mu_current, 0.2), arrowprops=dict(arrowstyle="->", lw=2.))
    ax2.set(title='likelihood(mu=%.2f) = %.2f\nlikelihood(mu=%.2f) = %.2f' % (mu_current, 1e14*likelihood_current, mu_new, 1e14*likelihood_new))
    
    # Posterior
    mu_post, sigma_post = call_posterior_analytical(data, mu_prior, sigma_prior)
    # Get the new parameters P(mu, sigma|data)
    posterior_analytical = norm(mu_post, np.sqrt(sigma_post)).pdf(x)
    posterior_current = norm(mu_post, np.sqrt(sigma_post)).pdf(mu_current)
    # get the likelihood of the current parameter (mu_current) given the posterior parameter
    posterior_new = norm(mu_post, np.sqrt(sigma_post)).pdf(mu_new)
    # get the likelihood of the new parameter (mu_new) given the posterior parameter
    ax3.plot(x, posterior_analytical)
    ax3.plot([mu_current] * 2, [0, posterior_current], marker='o', color='b')
    ax3.plot([mu_new] * 2, [0, posterior_new], marker='o', color=color)
    ax3.annotate("", xy=(mu_new, 0.2), xytext=(mu_current, 0.2),arrowprops=dict(arrowstyle="->", lw=2.))
    #x3.set(title=r'prior x likelihood $\propto$ posterior')
    ax3.set(title='posterior(mu=%.2f) = %.5f\nposterior(mu=%.2f) = %.5f' % (mu_current, posterior_current, mu_new, posterior_new))
    
    if accepted:
        mu_posterior.append(mu_new)
    else:
        mu_posterior.append(mu_current)
    ax4.plot(mu_posterior)
    ax4.set(xlabel='iteration', ylabel='mu', title='mu_posterior')
    plt.tight_layout()
    
"""
Note:
    If confused with plots. A Simple way to understand is to compare row 2 and row 1 or 
    row 3 and row 2 or similarly other rows. 
    1. If plot (2,1) has a green Vertical line and plot (1,1) has a red and blue Vertical line 
        -- then it means that the randomly chosen new parameter (mu_new) is more likely
           given the prior parameters (mu_prior and sigma_prior) 
           than the current parameters (mu_current).
    2. If plot (2,2) has a green Vertical curve and plot (1,2) has a red Vertical curve 
        -- then it means that the observed data is more likely given the new parameters 
           (mu_new and 1) when compared to the current parameter (mu_current, 1) 
           (here sigma_new is a constant)
    3. If plot (2,3) has a green Vertical line and plot (1,3) has a red and blue Vertical line 
        -- then it means that the randomly chosen new parameter (mu_new) is more likely
           given the posterior parameters (mu_post and sigma_post) 
           than the current parameters (mu_current).
    
           
"""
```

```python
# In reality we would not have conjugate scenario. So our posterior would simply be the multiple of the 
# likelihood and the prior. In MCMC we dont find the entire posterior distribution posterior distribution, 
# we just take random samples and jump to a parameter setting if the posterior for the current parameter in better'
# the posterior for the previous parameter setting.

# In the example we are using the parameters are the mean(mu) and the standard deviation (sigma). We would compute the
# likelihood and the prior for each parameter anf then evaluate if the posterior for the current parameter setting increaes.
# If so then we make a jump in the direction of the new parameter settings.

# In the following code we however assume a constant variance. Also to calculate the likelihood normally we should use
# the sum of log probabilities, but in this case for simplicity we just get the probability of each data point given the 
# parameters and then multiply them. In practise this may result in vanishing or exploding of parameters. So in practise
# we may use the maximum likelihood estimation to do the likelihood part.

def sample_parameters_chck_posterior(samples, data, mu_imit=0.5, proposal_width=0.5, mu_prior=0, sigma_prior=1, plot=False):
    mu_current = mu_imit
    mu_posterior = [mu_current]
    for i in np.arange(samples):
        # sample(Generate) a new mean
        mu_new = norm(mu_current, proposal_width).rvs()
#         print (mu_new)
        
        # Calculate the new likelihood, prior and posterior for the new parameter (mean).
        # norm(mu_current,1).pdf(data) will create the likelihood of all the data given the mean and prod()
        # will take the product of all the probability.
        likelihood_current = norm(mu_current,1).pdf(data).prod()
        likelihood_new = norm(mu_new,1).pdf(data).prod()
#         print(likelihood_current) # You will observe that the likelihood becomea a very small number approximating 1-e7
#         print(likelihood_new)
        
        # Calculating the priors
        prior_current = norm(mu_prior,sigma_prior).pdf(mu_current)  # Will product a simgle number because mu_current is just one value
        prior_new = norm(mu_prior,sigma_prior).pdf(mu_new)
        
        # Calculating the posterior without Normalization:
        posterior_current = prior_current * likelihood_current
        posterior_new = prior_new * likelihood_new
#         print(posterior_current) 
#         print(posterior_new)
        
        # Calculating the posterior probability.
        # If we divide the posterior_new be posterior_current the normalizing term in the denominator gets cancelled out
        # So for practical scenario we dont consider the normalizing term.
        posterior_prob = posterior_new/posterior_current
        # if the new posteriar is really bigger than than the posterior_current, this means that the data has bigger 
        # likelihood given the new parameter mu_new.
        
#         print ('posterior_prob ', posterior_prob)
        
        accept = np.random.rand() < posterior_prob
        
        # Only when the posterior probability is greater than a certain random
        # sample we jump to the mu_new else we dont. Note that we may have a 
        # sampling error. However thats acceptable because the overall method is computationally effective
        if accept:                 
#             print ('Iteration %d, Accepted mu_new !!'%i)  
            mu_current = mu_new    
        else:
#             print ('Iteration %d, Not Accepted mu_new !!'%i)
            pass
        if plot:
            plot_proposal(mu_current = mu_current,
                          mu_new = mu_new, 
                          mu_prior = mu_prior, 
                          sigma_prior = sigma_prior, 
                          data = observed_data, 
                          posterior_prob = posterior_prob,
                          accepted = accept,
                          mu_posterior = mu_posterior,
                          sample_no = i)
        
        # Now we store the posterior probability
        mu_posterior.append(mu_current)
    return mu_posterior

mu_posterior = sample_parameters_chck_posterior(5, data=observed_data, plot=True)
print (mu_posterior)
```

```python
# Let us now run take many-many iteration

num_iter = 10000
mu_posterior = sample_parameters_chck_posterior(num_iter, data=observed_data, plot=False)
print (len(mu_posterior))
print (len(np.arange(num_iter))+1)
ax = plt.subplot()
ax.plot(np.arange(num_iter+1), mu_posterior)
ax.set(xlabel='iteratition.. ', ylabel='posterior probability')
sns.despine()
```
