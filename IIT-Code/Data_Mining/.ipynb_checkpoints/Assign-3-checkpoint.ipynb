```python
# Data Mining Assignment 3:
"""
    Load the Boston dataset (sklearn.datasets.load boston()) 
    into Python us- ing a Pandas dataframe. Perform a K-Means analysis on unscaled data, 
    with the number of clusters ranging from 2 to 6. 
    Provide the Silhouette score to justify which value of k is optimal. 
    What information do the values of Homogene- ity/Completeness provide as well? 
    Calculate the mean values for all features in each cluster for the optimal clustering - 
    how do these values differ from the centroid coordinates?
"""


import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
from sklearn.cluster import KMeans
from sklearn import preprocessing
from sklearn import metrics

dataset = load_boston()
```

```python
# Load the Dataset into a Pandas DataFrame
df_dataset = pd.DataFrame(data=dataset.data,
                          columns=dataset.feature_names)
df_dataset.head()
```

```python
# Load the target into a pandas DataFrame
df_true_labels = pd.DataFrame(data=dataset.target, 
                         columns=['target'])
df_true_labels.head()
```

```python
# Scale the dataset and load it to a Pandas DataFrame
df_dataset_scaled = pd.DataFrame(data=preprocessing.scale(dataset.data),
                                 columns=dataset.feature_names)
df_dataset_scaled.head()
```

```python
# Get the Summary statistics of the oringinal dataset.
df_dataset.describe()
```

```python
# Get the Summary statistics of the scaled dataset.
df_dataset_scaled.describe()
```

```python
# Fit a K-means cluster
cluster_model_2 = KMeans(n_clusters=2, init='k-means++')
cluster_labels_2 = cluster_model_2.fit_predict(df_dataset)

cluster_model_3 = KMeans(n_clusters=3, init='k-means++')
cluster_labels_3 = cluster_model_3.fit_predict(df_dataset)

cluster_model_4 = KMeans(n_clusters=4, init='k-means++')
cluster_labels_4 = cluster_model_4.fit_predict(df_dataset)

cluster_model_5 = KMeans(n_clusters=5, init='k-means++')
cluster_labels_5 = cluster_model_5.fit_predict(df_dataset)

cluster_model_6 = KMeans(n_clusters=6, init='k-means++')
cluster_labels_6 = cluster_model_6.fit_predict(df_dataset)
```

```python
# Check the Silhouette Score of the Clusters, Silhouette scores portrays the closeness
# of the data points in a cluster to the seperation of the clusters. The more the Silhouette 
# score the better the clusters. The score ranges form -1 to 1
silhouette_score_2 = metrics.silhouette_score(df_dataset, cluster_labels_2)
silhouette_score_3 = metrics.silhouette_score(df_dataset, cluster_labels_3)
silhouette_score_4 = metrics.silhouette_score(df_dataset, cluster_labels_4)
silhouette_score_5 = metrics.silhouette_score(df_dataset, cluster_labels_5)
silhouette_score_6 = metrics.silhouette_score(df_dataset, cluster_labels_6)


print (silhouette_score_2)
print (silhouette_score_3)
print (silhouette_score_4)
print (silhouette_score_5)
print (silhouette_score_6)
```

```python
# Using Homogenity and Completeness score:
h_score2 = metrics.homogeneity_score(df_true_labels['target'],
                                    cluster_labels_2)
c_score2 = metrics.completeness_score(df_true_labels['target'],
                                     cluster_labels_2)

h_score3 = metrics.homogeneity_score(df_true_labels['target'],
                                    cluster_labels_3)
c_score3 = metrics.completeness_score(df_true_labels['target'],
                                     cluster_labels_3)

h_score4 = metrics.homogeneity_score(df_true_labels['target'],
                                    cluster_labels_4)
c_score4 = metrics.completeness_score(df_true_labels['target'],
                                     cluster_labels_4)

h_score5 = metrics.homogeneity_score(df_true_labels['target'],
                                    cluster_labels_5)
c_score5 = metrics.completeness_score(df_true_labels['target'],
                                     cluster_labels_5)

h_score6 = metrics.homogeneity_score(df_true_labels['target'],
                                    cluster_labels_6)
c_score6 = metrics.completeness_score(df_true_labels['target'],
                                     cluster_labels_6)

print (h_score2, c_score2)
print (h_score3, c_score3)
print (h_score4, c_score4)
print (h_score5, c_score5)
print (h_score6, c_score6)
```

```python
# Now that we know that the optimal clusters are formed when the no_of_clusters (k) = 3
# cluster_labels_3 contains the label of respective clusters number to with each row of the 
# dataframe (df_dataset) is assigned

# We will now add the label term to the dataframe:
df_dataset['class'] = cluster_labels_3
df_dataset.head()
```
