# Marginal and Conditional PDF for a High contrast subspace
#### This is called high contrast subspace because the marginal pdf of a **variable (var_0003)**-> indicated in grey is quite different than the conditional pdf of a ** variable (var_0003) ** --> indicated in pink.

A good way to understand this is to see conditional distribution, We containt the y value by some number, which means samples from y are taken using the constraint, then x values corresponding to all the samples y are collected and distribution is observed. Basically we tare testing if the marginal and conditional pdf for two variables come from the same distribution. If the t-stat is too small then we reject the null hypothesis that the two pdf comes from the same distribution

#### Marginal = dist (x) which can also be said as dist (x | all y's)
#### Conditional = dist(x | sampled y values using constraint)
#### if dist (marginal very differen form conditional , then the subspace is high contrast subspace)

### Marginal PDF: P(var_0002) and P(var_0003): Indicated in gray color:

```python
plt.figure(0, figsize=(20,20))
ax1 = plt.subplot2grid((3,3), (0,0), colspan=2)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2, rowspan=2,sharex=ax1)
ax3 = plt.subplot2grid((3,3), (1, 2), rowspan=2,sharey=ax2)
plt.suptitle("High Contrast Subspace: var_0002 & var_0003")

ax2.scatter(df['var_0003'],df['var_0002'])
df['var_0003'].hist(color='k', alpha=0.5, bins=30, ax=ax1, normed = True)
df['var_0002'].hist(color='k', alpha=0.5, bins=30, orientation="horizontal")
```

### Conditional PDF (P(var_0003) | points_rank(var_0002) >0.65))  -> Indicated in pink color

```python
plt.figure(0, figsize=(20,20))
ax1 = plt.subplot2grid((3,3), (0,0), colspan=2)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2, rowspan=2,sharex=ax1)
ax3 = plt.subplot2grid((3,3), (1, 2), rowspan=2,sharey=ax2)
plt.suptitle("High Contrast Subspace: var_0002 & var_0003")
ax2.scatter(df['var_0003'],df['var_0002'])
df['var_0003'].hist(color='k', alpha=0.5, bins=30,ax=ax1,normed = True)
df['var_0002'].hist(color='k', alpha=0.5, bins=30,orientation="horizontal")

var2_rank = df['var_0002'].rank() # Will simply rank by the value of the column var_0002
var2_maxRank = max(df['var_0002'].rank())
var0002_index = var2_rank/var2_maxRank   # min still remains minimum and maximum still remains maximum, its just done to bound the values by (0,1)
# Find the indices data points that have rank more than 0.65 and get only the column var_003
dataPoints = df[(var0002_index>0.20) & (var0002_index<0.80)]['var_0003']
print (dataPoints.shape)

# Plot the conditional histogram
# The below is similar to finding the distribution of var_0003 given the data points that have var_0002>0.65
dataPoints.hist(color='r', alpha=0.2, bins=20,ax=ax1,normed = True,linestyle='dotted')
# ax1.text(0.8,5,'p-value:<0.0001***',va="center", ha="center")
plt.show()
```

#### Testing the difference between the conditional and marginal pdf's

```python
t=stats.ttest_ind(df['var_0003'].values, df[(var0002_index>0.65)]['var_0003'].values, equal_var = False)
print ("t-test")
print (t.pvalue)
```

# Marginal and Conditional PDF for a LOW contrast subspace:
#### This is called low contrast subspace because the marginal pdf of a **variable (var_0003)**-> indicated in grey is quite similar to the conditional pdf of a ** variable (var_0003) ** --> indicated in pink.

Basically we are testing if the marginal and conditional pdf for two variables come from the same distribution. If the t-stat is too large then we forbid to reject the null hypothesis that the two pdf comes from the same distribution

### Marginal PDF: P(var_0000) and P(var_0003): Indicated in gray color:

```python
plt.figure(0, figsize=(20,20))
ax1 = plt.subplot2grid((3,3), (0,0), colspan=2)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2, rowspan=2,sharex=ax1)
ax3 = plt.subplot2grid((3,3), (1, 2), rowspan=2,sharey=ax2)
plt.suptitle("High Contrast Subspace: var_0002 & var_0003")

ax2.scatter(df['var_0000'],df['var_0003'])
df['var_0000'].hist(color='k', alpha=0.5, bins=30,ax=ax1,normed = True)
df['var_0003'].hist(color='k', alpha=0.5, bins=30,orientation="horizontal")
```

### Conditional PDF (P(var_0003) | points_rank(var_0000) >0.65))  -> Indicated in pink color

```python
plt.figure(0, figsize=(20,20))
ax1 = plt.subplot2grid((3,3), (0,0), colspan=2)
ax2 = plt.subplot2grid((3,3), (1,0), colspan=2, rowspan=2,sharex=ax1)
ax3 = plt.subplot2grid((3,3), (1, 2), rowspan=2,sharey=ax2)
plt.suptitle("High Contrast Subspace: var_0002 & var_0003")

ax2.scatter(df['var_0000'],df['var_0003'])
df['var_0000'].hist(color='k', alpha=0.5, bins=30,ax=ax1,normed = True)
df['var_0003'].hist(color='k', alpha=0.5, bins=30,orientation="horizontal")

var0003_index = df['var_0003'].rank()/max(df['var_0003'].rank())
dataPoints = df[(var0003_index > 0.65)]['var_0000']

dataPoints.hist(color='r', alpha=0.2, bins=20,ax=ax1,normed = True,linestyle='dotted')
# ax1.text(0.8,5,'p-value:<0.0001***',va="center", ha="center")
plt.show()
```

```python
t=stats.ttest_ind(df['var_0000'].values, df[(var0003_index>0.65)]['var_0000'].values, equal_var = False)
print ("t-test")
print (t.pvalue)
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from scipy.io import arff
import pandas as pd
from scipy import stats
import itertools
import random
import numpy as np
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import MinMaxScaler
from sklearn import metrics
```

```python
index_df = (df.rank()/df.rank().max()).iloc[:,:-1]   # Remove the class column
print (index_df.head())
```

```python
def columnPairGenerator(startPoint,space,n):
    columnPair=[]
    for item in itertools.combinations(list(set(space)-set(startPoint)),(n-len(startPoint))):
        columnPair.append(sorted(startPoint+list(item)))
    return columnPair
```

```python
listOfcolumnPair = columnPairGenerator([],df.columns[:-1],2)
print (listOfcolumnPair)
testedColumnPair=[]
selection=[]
```

```python
i=0
while len(listOfcolumnPair)>0:
    if listOfcolumnPair[0] not in testedColumnPair:
        columnPair = listOfcolumnPair[0]
#         print ('')
#         print ('')
#         print ('Running for column pair : ', columnPair)
#         print ('')
        alpha1 = pow(0.2,(float(1)/float(len(columnPair))))   # listOfColumnPair[0] = number of dimension in that list (listOfcolumnPair)
#         print ('alpha1 ', alpha1)
        pvalue_Total =0
        pvalue_cnt = 0
        avg_pvalue=0

        # We iterate for 50 random conditional and marginal distribution
        for i in range(0,50):
            lband = random.random()
            uband = lband+alpha1
            v = random.randint(0,(len(columnPair)-1))              # is the marginal column
#             print ('lband, uband, v = ', lband, uband, v)
#             print ('The marginal column is: ', columnPair[v])
            rest = list(set(columnPair)-set([columnPair[v]]))      # are all teh conditional columns
#             print ('The conditional columns are ', rest)                       
#             print (df[columnPair[v]].values.shape)
#             print (df[((index_df[rest]<uband) & (index_df[rest]>lband))].shape)
#             print (df[((index_df[rest]<uband) & (index_df[rest]>lband)).all(axis=1)].shape)
#             print (columnPair[v])
#             print (df[((index_df[rest]<uband) & (index_df[rest]>lband)).all(axis=1)][columnPair[v]].values.shape)
            k = stats.ks_2samp(df[columnPair[v]].values,                 # v -> The column (marginal column)       
                               df[((index_df[rest]<uband) &              # rest -> The conditinoal columns 
                                   (index_df[rest]>lband)).all(axis=1)
                               ][columnPair[v]].values)
            
#             print ('The statistic is : ', k.pvalue)
            if not(np.isnan(k.pvalue)):
                pvalue_Total = pvalue_Total+k.pvalue
                pvalue_cnt = pvalue_cnt+1
        
        if pvalue_cnt>0:
            avg_pvalue = pvalue_Total/pvalue_cnt
            
#         print (avg_pvalue)
        
        if (1.0-avg_pvalue)>0.6:
            selection.append(columnPair)
            listOfcolumnPair = listOfcolumnPair + columnPairGenerator(columnPair,df.columns[:-1],(len(columnPair)+1))
        
#         print (listOfcolumnPair)
            
        testedColumnPair.append(listOfcolumnPair[0])
        listOfcolumnPair.pop(0)
        listOfcolumnPair = [list(t) for t in set(map(tuple,listOfcolumnPair))]
#         print ('67676767676767676 ', listOfcolumnPair)
    else:
        listOfcolumnPair.pop(0)

selection
```

```python
def knn(df,k):
    nbrs = NearestNeighbors(n_neighbors=3)
    nbrs.fit(df)
    distances, indices = nbrs.kneighbors(df)
    return distances, indices

def reachDist(df,MinPts,knnDist):
    nbrs = NearestNeighbors(n_neighbors=MinPts)
    nbrs.fit(df)
    distancesMinPts, indicesMinPts = nbrs.kneighbors(df)
    distancesMinPts[:,0] = np.amax(distancesMinPts,axis=1)
    distancesMinPts[:,1] = np.amax(distancesMinPts,axis=1)
    distancesMinPts[:,2] = np.amax(distancesMinPts,axis=1)
    return distancesMinPts, indicesMinPts

def ird(MinPts,knnDistMinPts):
    return (MinPts/np.sum(knnDistMinPts,axis=1))

def lof(Ird,MinPts,dsts):
    lof=[]
    for item in dsts:
        tempIrd = np.divide(Ird[item[1:]],Ird[item[0]])
        lof.append(tempIrd.sum()/MinPts)
    return lof
```
