```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_boston
from sklearn.cluster import KMeans

from sklearn import preprocessing
from sklearn import metrics
```

```python
%matplotlib inline
```

```python
dataset = load_boston()
df = pd.DataFrame(data=dataset.data,
                  columns=dataset.feature_names)
df_scaled = pd.DataFrame(data=preprocessing.scale(dataset.data),
                         columns=dataset.feature_names)
```

```python
sns.clustermap(df,
               method='average',
               metric='euclidean',
               z_score=None,
               standard_scale=None)
```

```python
clust_model = KMeans(n_clusters=4,
                     init='k-means++')
clust_labels = clust_model.fit_predict(df)
print(clust_labels)
```

```python
silhouette_avg = metrics.silhouette_score(df,
                                          clust_labels)

print(silhouette_avg)
```

```python
target_labels = dataset.target
#target_labels = scale(dataset.target)

h_score = metrics.homogeneity_score(target_labels,
                                    clust_labels)
c_score = metrics.completeness_score(target_labels,
                                     clust_labels)
r_score = metrics.adjusted_rand_score(target_labels,
                                      clust_labels)
mi_score = metrics.adjusted_mutual_info_score(target_labels,
                                              clust_labels)

print(h_score,
      c_score,
      r_score,
      mi_score)
```

```python
from sklearn.decomposition import PCA

pca_col = PCA(2).fit_transform(df)
plt.scatter(x=pca_col[:,0],
            y=pca_col[:,1],
            c=clust_labels)
plt.show()
```
