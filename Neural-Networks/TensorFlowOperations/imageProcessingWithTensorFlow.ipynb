# Convolution with Gaussian Kernel
-------

#### Creating a Gaussian Kernel 1D

```python
# Creating gaussian plots with the linspace 
# Writing the formula explicitely using multiple tensor operations

mean = 0
sigma = 1.0

gaussianFunctionTF = 1/(sigma* tf.sqrt(2.0*3.14)) * tf.exp((tf.negative(tf.pow((ls-mean),2) / (2.0 * tf.pow(sigma, 2)))))

gaussianFunction = gaussianFunctionTF.eval()
plt.plot(gaussianFunction)
```

#### Creating a Gaussian Kernel 2D

```python
# creating a 2D gaussian kernel
arrSize = gaussianFunctionTF.get_shape().as_list()[0]

gaussianFunctionTF2D = tf.matmul(tf.reshape(gaussianFunction, (arrSize,1)), tf.reshape(gaussianFunctionTF, (1, arrSize)))

gaussianFunction2D = gaussianFunctionTF2D.eval()
plt.imshow(gaussianFunction2D)
```

#### Apply convolution with gaussian kernel to a image

```python
# Load a dataset from sckit learn
from skimage import data

cameraMan = data.camera().astype('float32')
plt.imshow(cameraMan, cmap='gray')
print (cameraMan.shape)
```

```python
# In tensor flow images are passed in four tensors
# 1. batchSize
# 2. height
# 3. width
# 4. Channels
# In order to perform a convolution with gaussian Kernel in tensor flow we need to have our image in the same dimension in Tensor flow.
# Let us convert them.
imageTF = tf.reshape(cameraMan, [1, cameraMan.shape[0], cameraMan.shape[1], 1])
print (imageTF.get_shape().as_list())

# The input of the gaussian kernel should again be a 4D tensor
# 1. height of the kernel
# 2. width of the kernel
# 3. Channels
# 4. Kernels
gaussianKernelTF = tf.reshape(gaussianFunction2D, [gaussianFunction2D.shape[0], gaussianFunction2D.shape[1], 1, 1])
print (gaussianKernelTF.get_shape().as_list())
```

```python
# Now we apply the convolution to the image using tensor opereation
convImageTF = tf.nn.conv2d(imageTF,gaussianKernelTF, strides=[1,1,1,1], padding='SAME')
print (convImageTF.get_shape().as_list())
convImage = convImageTF.eval()
# the strides indicates how much you want to shift the kernel for the consequent convolution
# [1,1,1,1,] indicates [number_of_images, height_of_image, width_of_image, num_channels] 
# padding SAME says that the image size would not decrease (no down-sampling) after convolution as
# extra padding are added in all 4 sides of the image to avoid depletion in image size (down sampling).
```

```python
# plotting the images
# Tensor objects have 4 tensors the plotting 
# We discuss two ways 1) by squeezing the image, 2) manually specifying the dimension 
fig, axs = plt.subplots(1,2, facecolor='y', edgecolor='k')
axs[0].imshow(np.squeeze(convImage), cmap='gray')
axs[1].imshow(convImage[0,:,:,0])
```

# Convolution with Gabor kernel
-------

Garbor Kernels are the combination of a gaussian Kernel and sinewave. 

#### Creating a Gabor Kernel 1D

```python
sinwaveFunctionTF = tf.sin(ls)
sinwaveFunction = sinwaveFunctionTF.eval()

# Gabor Filter | Gabor Kernel --> Can be thought as bimodal, more narrow
# To calculate the Gabor, we need to multiply each corresponsing point
# of the gaussian function with the sine function
gaborFunctionTF = tf.multiply(gaussianFunctionTF, sinwaveFunctionTF)
print (gaborFunctionTF.get_shape().as_list())
gaborFunction = gaborFunctionTF.eval()


# Plot Gaussian | Sine | Gabor
fig, axs = plt.subplots(1,3, figsize=(15,5),facecolor='y', edgecolor='k')
axs[0].plot(gaussianFunction)
axs[0].set_title('GAUSSIAN')
axs[1].plot(sinwaveFunction)
axs[1].set_title('SINE')
axs[2].plot(gaborFunction)
axs[2].set_title('GABOR')
```

#### Creating a Gabor Kernel 2D

```python
# First we create a 2D matrix for sine wave and the gaussian 
# and then we compute the matrix multiply (matrix dot product)

sineFunctionTF2D = tf.matmul(tf.reshape(sinwaveFunctionTF, [arrSize, 1]), tf.ones((1, arrSize)))
print (sineFunctionTF2D.get_shape().as_list())
# We use ones because the dot product of two gaussians produces a gaussian but the same may not
# be apply for sinwave fucnction. So inorder to preserve the sinwave function at every dimension
# we take the dot product of the sine wave with ones

gaborFunctionTF2D = tf.multiply(gaussianFunctionTF2D, sineFunctionTF2D)
print (gaborFunctionTF2D.get_shape().as_list())

# # Plot Gaussian | Sine | Gabor
fig, axs = plt.subplots(1,3, figsize=(15,5),facecolor='y', edgecolor='k')
axs[0].imshow(gaussianFunctionTF2D.eval())
axs[0].set_title('GAUSSIAN')
axs[1].imshow(sineFunctionTF2D.eval())
axs[1].set_title('SINE')
axs[2].imshow(gaborFunctionTF2D.eval())
axs[2].set_title('GABOR')
```

#### Apply convolutions with Gabor filter to the image

We build a new graph usign for the convolution with gabor kernel. We use placeholders and build lineage of graphs to be executed in chain

```python
# Here we make the use of the tensor flow placeholders and dynamically put the image
imageTF_2D = tf.placeholder(tf.float32, shape=[None, None], name='img_CameraMan')

# The tensor above has shape of 2 tensor,
# We know to apply convolutions we need a 4D Tensor shape of the variable
# Below we extend the dimension of the above image using the expand_dims function
imageTF_3D = tf.expand_dims(imageTF_2D, 2) # This extends the 3rd dimension. It may result in [None, None, 1]

imageTF_4D = tf.expand_dims(imageTF_3D, 0) # This extends the 1st dimension. It may result in [1, None, None, 1]

print (imageTF_4D.get_shape().as_list())
```

```python
# We build Tensor placeholder for the gabor parameteres like mean, sigma, 
# and perform the convolution
gaussianMeanTF = tf.placeholder(tf.float32, name='GaborMean')
gaussianSigmaTF = tf.placeholder(tf.float32, name='GaborSigma')
kernelSizeTF = tf.placeholder(tf.int32, name='KernelSize')

weights = tf.linspace(-3.0,3.0,kernelSizeTF)

# Building the Gaussian Kernel Dynamically:
gaussian1D_TF = 1/(gaussianSigmaTF* tf.sqrt(2.0*3.14)) * tf.exp((tf.negative(tf.pow((weights-gaussianMeanTF),2) / (2.0 * tf.pow(gaussianSigmaTF, 2)))))
gaussian2D_TF = tf.matmul(
                    tf.reshape(gaussian1D_TF, tf.stack([kernelSizeTF, 1])),
                    tf.reshape(gaussian1D_TF, tf.stack([1, kernelSizeTF]))
                )

# Buildign the SineWave Kernel Dynamically
sinewave1D_TF = tf.sin(weights)
sinewave2D_TF = tf.multiply(
                    tf.reshape(sinewave1D_TF, tf.stack([kernelSizeTF, 1])),
                    tf.ones(tf.stack([1,kernelSizeTF])))

# Build the Gabor Kernel
gabor2D_TF = tf.multiply(gaussian2D_TF, sinewave2D_TF)
gabor4D_TF = tf.reshape(gabor2D_TF, shape=[kernelSizeTF, kernelSizeTF, 1, 1])  # 1,1 -> channel, num_of_kernel
```

```python
# Now we feed the values to the placehoders in the graph and compute the entire graph
convImageGabor_TF = tf.nn.conv2d(imageTF_4D, gabor4D_TF, strides=[1,1,1,1], padding='SAME', name='gaborConvolution')

# Finally we get the convoluted image
convolutedImage = convImageGabor_TF[0,:,:,0] # 0,0 because we are not interested in the batch size and the number of channels
```

```python
# Compute the complete graph
feedDict = {
    imageTF_2D: cameraMan,
    gaussianMeanTF: 0,
    gaussianSigmaTF: 0.5,
    kernelSizeTF: 32
}

res = convolutedImage.eval(feed_dict = feedDict)
```

```python
# Plot the convoluted image
print (res.shape)
plt.imshow(res, cmap='gray')
```
