Great Circle Distance: Is the shortest distance between two points in a sphere. An euclidean distance is computed using a straight line, however, we can not use euclidean distance to compute the distance between two points in a sphere. As a matter of fact the two points separate the sphere in two archs. The smaller arch is the shortest distance between the points.

** Problems: ** When two points in the surface are very very close then the, let's say "1 km" apart and if the system has low floating point precision, then the consine angle between them comes out to be 0.99999. However, for a modern 64 bit floating point system the distance between very close points (say few meters) doesnt pose much problem. However, it is always good to use ** Haversine distance ** is found to be better for very small distances. Despite, Haversine distance is best suited for near points and serves as an obvious distance measure, for some points such as antipodal points (points opposite to one another in the sphere). Another good distance based algorithm is the ** Vincenty Formula ** for an ellipsoid with equal major and minor axes.



*** Haversine Formula ***

Lets say, we have two points with lon1, lat1 and lon2, lat2

* Earth Radius R: 6,371 km (3,959 mi)
* Packages in R use : 6378.388 km

Formula: 
$$2* R * arcsine \sqrt{sin^2(\frac{lat1-lat2} 2 ) + cos(lat1) * cos(lat2) * sin^2(\frac{lon1-lon2} 2)} $$

For Programs:

$$2* R * asin * [ min(1, \sqrt{sin^2(\frac{lat1-lat2} 2 ) + cos(lat1) * cos(lat2) * sin^2(\frac{lon1-lon2} 2)})] $$
